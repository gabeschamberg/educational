The quiz will be 15 minutes. Do your best, and if you don't know the answer, explain what you do know!


A snowboarding goat wants to learn how ride a snowboard in the halfpipe. He decides it would be safest to first do a computer simulation. The goat decides that for his simulation, the shape of the hafpipe will be given by the equation:

\begin{equation}
f(x,y)=2y^2-5x+100
\end{equation}

1) For $x=5$, sketch $f(x,y)$. In other words, sketch $f(5,y)$ as a function of $y$ (in 2 dimensions)

2) For $y=0$, sketch $f(x,y)$. In other words, sketch $f(x,0)$ as a function of $x$ (in 2 dimensions)

3) Find the partial derivatives $\frac{\partial f}{\partial x}$ and $\frac{\partial f}{\partial y}$.

4) In general, you can find a local min/max point by setting both partial derivatives equal to zero. Can we do that for this equation? Why or why not?

# Solutions

```python
# numpy for vector operations
import numpy as np

# matplotlib for plotting
%matplotlib inline
import matplotlib.pyplot as plt
from matplotlib import cm

from mpl_toolkits.mplot3d import Axes3D
```

```python
def halfpipe(x,y):
    return 2*(y**2) + 100 - 5*x

fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')
x = np.arange(0,20,0.05)
y = np.arange(-10, 10, 0.05)
X, Y = np.meshgrid(x, y)

zs = np.array([halfpipe(x,y) for x,y in zip(np.ravel(X), np.ravel(Y))])
Z = zs.reshape(X.shape)

ax.plot_surface(X, Y, Z,cmap = cm.cool,linewidth = 0.2)


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

plt.show()
```

```python
fig = plt.figure(figsize=(15,4))
ax = fig.add_subplot(121, projection='3d')
x = np.arange(0,20,0.05)
y = np.arange(-10, 10, 0.05)
X, Y = np.meshgrid(x, y)

for i in range(-10,10):
    z = halfpipe(x,i)
    if(i==0):
        ax.plot(x,len(z)*[i], z,'r',linewidth = 1)
    else:
        ax.plot(x,len(z)*[i], z,'b',linewidth = 1)


ax.set_xlabel('X')
ax.set_ylabel('Y')
ax.set_zlabel('Z')

ax = fig.add_subplot(122)
x = np.arange(0,20,0.05)
y = np.arange(-10, 10, 0.05)
X, Y = np.meshgrid(x, y)

for i in range(-10,10):
    z = halfpipe(x,i)
    if(i==0):
        ax.plot(x,z,'r',linewidth = 1)
    else:
        ax.plot(x,z,'b',linewidth = 1)


ax.set_xlabel('X')
ax.set_ylabel('Z')

plt.show()
```

A goat is climbing a mountain made of loose rocks. Every couple of steps, the rocks slip and he slides back down a little bit. The vertical position (in meters) of the goat is given as a function of time $t$ (in seconds):

$$
y(t) = t + 2\sin\left(\frac{2\pi}{4}t\right)
$$

**1)** Give an expression for the goat's velocity in the vertical direction, $v(t)$? (include units) 

**2)** Give an expression for the goat's acceleration, $a(t)$? 

**3)** What is the maximum velocity, $v_{max}$? 

**4)** What is the acceleration when velocity is $v_{max}$? (i.e. if $v(t_{max})=v_{max}$, what is $a(t_{max})$)? 
 
**Bonus Question**: Explain why it makes sense that $\frac{d}{dx} \sin(x) = \cos(x)$ using words and/or pictures. ("I have it memorized" doesn't count!)

```python
import matplotlib.pyplot as plt
import numpy as np
%matplotlib inline

t = np.arange(0,30,0.1)
y = t + 2*np.sin((2*np.pi/4)*t)
plt.plot(t,y)
plt.xlabel('Time (s)')
plt.ylabel('Height (m)')
plt.title('Problem 1')
plt.show()

y = 1 + (2*np.pi/4)*2*np.cos((2*np.pi/4)*t)
plt.plot(t,y)
plt.plot(t,[0]*len(t),'--g')
plt.xlabel('Time (s)')
plt.ylabel('Velocity (m/s)')
plt.title('Problem 2')
plt.show()

y =-(2*np.pi/4)*(2*np.pi/4)*2*np.sin((2*np.pi/4)*t)
plt.plot(t,y)
plt.plot(t,[0]*len(t),'--g')
plt.xlabel('Time (s)')
plt.ylabel('Acceleration ((m/s)/s)')
plt.title('Problem 3')
plt.show()

x = np.arange(0,3,0.01)
plt.plot(x,np.cos(2*np.pi*x),'--g')
plt.plot(x,np.sin(2*np.pi*x))
plt.xlabel('x')
plt.legend(['cos(2*pi*x)' , 'sin(2*pi*x)'])
plt.title('Bonus')
plt.grid()
plt.show()
```
